```python
# Question 1: Create a New Database and  Table for Employees
#Task: Create a new database named company_db and Create a table named employees with the following columns:

Create database company_db;

use company_db;

create Table employees
(
employee_id int PRIMARY KEY,
first_name varchar(50),
last_name varchar(50),
department varchar(50),
salary int,
hire_date date
);
```

```python
# Question 3: Display All Employee Records Sorted by Salary (Lowest to Highest)

Select * from employees
ORDER BY salary ASC;
```

```python
# Question 4: Show Employees Sorted by Department (AZ) and Salary (High  Low)

Select * from employees
ORDER BY department ASC,
salary DESC;
```

```python
# Question 5: List All Employees in the IT Department, Ordered by Hire Date (Newest First)

Select * from employees
WHERE department = 'IT'
ORDER BY hire_date DESC;
```

```python
# Question 9: What is the Difference Between a Primary Key and a Foreign Key?

Answer:
In databases, Primary Key and Foreign Key are both used to maintain data integritybut they serve different purposes.

Primary Key: A Primary Key uniquely identifies each row in a table.
- Must be unique
- Cannot be NULL
- Only one primary key per table (can be composite)
- Ensures entity integrity

Foreign Key: A Foreign Key is a column (or set of columns) in one table that refers to the Primary Key in another table.
- Can have duplicate values
- Can be NULL
- A table can have multiple foreign keys
- Maintains referential integrity
```

```python
# Question 10: What Are Constraints in SQL and Why Are They Used?

Answer:
In SQL, constraints are rules applied to table columns to restrict what data can be stored and to maintain data accuracy and integrity. They prevent invalid data from being inserted, updated, or deleted.

Constraints are used for:
- Ensure correct data entry
- Prevent duplicates
- Maintain relationships between tables
- Enforce business rules
- Improve data reliability
```
